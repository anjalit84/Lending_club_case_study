### Importing Libraries

```python
import pandas as pd
import numpy as np
```

### Importing Dataset

```python
raw_input=pd.read_csv('loan.csv')

#Loading_column names
raw_columns=raw_input.columns
len(raw_columns)
```

#### Checking default length

```python
raw_length=len(raw_input)
raw_length
```

# Cleaning the data set

### Checking for null values

Creating a DF which can hold raw column information. Starting with creating a column in DF to check for null values

```python
#Function to calculate percent of column missing
def function(x):
    return (0 if x==0 else (x/raw_length)*100)
```

```python
#checking number of missing values in each column
raw_missing=list(raw_input.isnull().sum())

#Calculating percent of missing values
raw_percent_missing=list(map(function,raw_missing))
```

```python
#creating dataframe

column_analysis_summary = pd.DataFrame(list(zip(raw_columns,raw_missing,raw_percent_missing)) , columns =['Column Name', 'Number of missing','Percent Missing'])
column_analysis_summary['Type']=raw_input.dtypes.tolist()
column_analysis_summary
```

```python
#Splitting columns where missing values is 100% vs columns which have values (Only for Analysis)
column_analysis_summary["Removed"] = "No"
column_analysis_summary.loc[column_analysis_summary["Percent Missing"] == 100, "Removed"] = "Yes"
column_analysis_summary
```

### Data Splitting based on columns

Creating 2 data frames - 
one where there is no missing data in columns, 
Second where there is missing values but not completly null

```python
#Dataframe with partial missing values - 14 columns
partial_missing_raw=raw_input[column_analysis_summary[(column_analysis_summary['Percent Missing']!=100) & (column_analysis_summary['Percent Missing']!=0)]['Column Name'].tolist()]

#Dataframe with no missing values - 43 columns
complete_raw=raw_input[column_analysis_summary[(column_analysis_summary['Percent Missing']!=100) & (column_analysis_summary['Percent Missing']==0)]['Column Name'].tolist()]
```

```python
column_analysis_summary[(column_analysis_summary['Percent Missing']!=100) & (column_analysis_summary['Percent Missing']!=0)]
```

```python
#dropping mths_since_last_record and next_pymnt_d as the number of missing values is very high + Removing desc as the values are randomly entered by user(Can be processed later)
partial_missing_raw=partial_missing_raw.drop(['mths_since_last_record','next_pymnt_d'], axis=1).copy()
column_analysis_summary.loc[column_analysis_summary["Column Name"] == 'next_pymnt_d', "Removed"] = "Yes"
column_analysis_summary.loc[column_analysis_summary["Column Name"] == 'mths_since_last_record', "Removed"] = "Yes"
column_analysis_summary.loc[column_analysis_summary["Column Name"] == 'desc', "Removed"] = "Yes"
column_analysis_summary[(column_analysis_summary['Percent Missing']!=100) & (column_analysis_summary['Percent Missing']!=0)]
```

```python
#combine full and partial dataframes
transform_raw=pd.concat([complete_raw, partial_missing_raw], axis=1)
```

## Creating columns in columns df to analze unique values

```python
column_analysis_summary[['Unique_values_count','Unique_values_list']]=column_analysis_summary.apply(lambda x: pd.Series([transform_raw[x['Column Name']].nunique(),transform_raw[x['Column Name']].unique()]) if x['Removed']=='No' else (-9999,-9999), axis=1)
```

    Removing columns that have only one unique value and all unique values(id,member_id and url)

```python
column_analysis_summary.loc[((column_analysis_summary["Unique_values_count"] == 1) | (column_analysis_summary["Unique_values_count"] == raw_length)), "Removed"] = "Yes"
transform_raw=transform_raw.drop((list(column_analysis_summary[((column_analysis_summary["Unique_values_count"] == 1) | (column_analysis_summary["Unique_values_count"] == raw_length))]['Column Name'])), axis=1).copy()
```

```python
column_analysis_summary[column_analysis_summary["Removed"] == 'No']
```

```python
#Remove '%' and convert to float for int_rate & revol_util

transform_raw['int_rate']=(transform_raw['int_rate'].apply(lambda x: format(float(x.split("%")[0]), ".2f"))).astype(float)
transform_raw['revol_util']=(transform_raw['revol_util'].apply(lambda x: format(float(x.split("%")[0]), ".2f") if pd.notnull(x) else x)).astype(float)
column_analysis_summary.loc[column_analysis_summary["Column Name"] == 'int_rate', "Type"] = "float64"
column_analysis_summary.loc[column_analysis_summary["Column Name"] == 'revol_util', "Type"] = "float64"

#Split date column to date and month for issue_d,earliest_cr_line,last_pymnt_d,last_credit_pull_d and delete the original column

transform_raw[['issue_d_month','issue_d_date']] = transform_raw.issue_d.str.split("-",expand=True)
transform_raw[['earliest_cr_line_month','earliest_cr_line_date']] = transform_raw.earliest_cr_line.str.split("-",expand=True)
transform_raw[['last_pymnt_d_month','last_pymnt_d_date']] = transform_raw.last_pymnt_d.str.split("-",expand=True)
transform_raw[['last_credit_pull_d_month','last_credit_pull_d_date']] = transform_raw.last_credit_pull_d.str.split("-",expand=True)

#deleting trasnformed date columns and title col as well as the text is random
transform_raw=transform_raw.drop(['issue_d','earliest_cr_line','title','last_pymnt_d','last_credit_pull_d'], axis=1).copy()
column_analysis_summary.loc[column_analysis_summary["Column Name"] in ['issue_d','earliest_cr_line','title','last_pymnt_d','last_credit_pull_d'], "Removed"] = "Yes"

#splitting zipcode and transforming columns type
transform_raw['pub_rec_bankruptcies']=transform_raw['pub_rec_bankruptcies'].astype(object)
transform_raw['zip_code']=transform_raw['zip_code'].apply(lambda x: x.split('x')[0])
column_analysis_summary.loc[column_analysis_summary["Column Name"] == 'pub_rec_bankruptcies', "Type"] = "object"
```

```python
column_analysis_summary["Min_Max"] = "None"
for i in column_analysis_summary['Column Name']:
    if (i in transform_raw.columns) and (transform_raw[i].dtype =='int64' or transform_raw[i].dtype =='float64'):
        column_analysis_summary.loc[column_analysis_summary["Column Name"] == i, "Min_Max"] = str([transform_raw[i].min(),transform_raw[i].max()])
```
